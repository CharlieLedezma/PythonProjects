```python
import numpy as np
import matplotlib.pyplot as plt
import control as co
```

```python
GA = co.tf([1],[1, 2, 1])
GA
```

```python
GB = co.tf([1], [1, 0.4, 1])
GB
```

```python
GC = co.tf([1], [1, 5, 1])
GC
```

```python
GD = co.tf([1],[1,1,1])
GD
```

```python
t= np.linspace(0,10, 1000)
t, y = co.step_response(GA)
plt.plot(t,y)
plt.xlabel('time(s)')
plt.ylabel('Amplitude')
plt.title('Impulse Response')
plt.grid()
print("O ganho DC é de {}".format(co.dcgain(GA)))
print("O Rise time é de {} segundos, o Settling time é {} segundos e o Overshoot é de {}".format(co.step_info(GA)['RiseTime'], co.step_info(GA)['SettlingTime'], co.step_info(GA)['Overshoot']))
```
