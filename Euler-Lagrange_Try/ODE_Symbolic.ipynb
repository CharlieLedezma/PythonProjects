```python
# %% Symbolic library and its initialization
import sympy as sy
# For function plot
from sympy.plotting import plot
```

```python
# %% Some common variables
x = sy.symbols('x')
y = sy.Function('y')
y1 = sy.Derivative(y(x), x)
```

```python
# Definition
eqdiff = y1 + (y(x)**2)/x**2
```

```python
# Solution
sol = sy.dsolve(eqdiff, y(x))

# Print
sy.pprint(sol)
```

```python
# %% Solution with the initial conditions
sol = sy.dsolve(eqdiff, y(x),ics={y(1): '1/2'})
sy.pprint(sol)
```

```python
# %% Solution with the initial conditions
sol = sy.dsolve(eqdiff, y(x),ics={y(1): '1/2'})
sy.pprint(sol)

# %% Plot
title = sy.latex(sol.rhs)
xtol = 1e-3
p1 = plot(sol.rhs, (x,-1,1/3-xtol), show=False, ylim=[-10,10], title=f"${title}$")
p2 = plot(sol.rhs, (x,1/3+xtol,1), show=False, ylim=[-10,10])
p1.extend(p2)
p1.show()

# %% Save in svg format
p1.save("ode-1.svg")
```
